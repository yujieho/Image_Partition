## Image partitions / color reducing using diffusion maps

### 1. Import some libraries for programming

```python
%pylab inline
import cv2
from itertools import product
from scipy import sparse
from scipy.sparse.linalg import svds
from scipy.spatial.distance import cdist
from sklearn.cluster import KMeans
```

### 2. Load image

```python
img = cv2.imread('data/flower.jpg')
img = img[:,:,::-1]

#normalize the RGB code
img = img/255

#show image
plt.imshow(img)
plt.show()

#size
print('Size of figure: ', img.shape[0],'*',img.shape[1])
```

```python
#create a down-sampled version of image
scale = 20

imc = img[::scale,::scale,:]
imc = zeros(imc.shape)

for i,j in product(range(scale),range(scale)):
    imc += img[i::scale,j::scale,:]
imc = imc/(scale**2)

#show image
plt.imshow(imc)
plt.show()

#size
x, y, z = imc.shape
n = x*y
print('Size of figure: ', x,'*',y)
```

### 3. Define Functions

```python
def diffuse(datav, clusters, dim):    
    sigma, dist = getSnG(datav)
    K = getK(sigma, dist)
    D, Q = getQ(K)
    S, psi = get_eig(Q, D)
    Y = getY(S, psi)
    labels = getlabels(Y)
    
    return labels
```

```python
#sigma is an n*n sparse matrix with local scaling parameters on its diagnoal.
#dist is an n*n symmetric sparse matrix with entries of distance between close datas.

def getSnG(datav):

    lsigma=[]
    ldistptr=[0]
    ldistind=[]
    ldist=[]

    datavs=[]
    for i in range(n):
        temp=cdist([datav[i]],datav)
        threshold=sum(temp)/n    #sum(temp)/n, percentile(temp,20)

        ldist.extend(temp[temp<threshold]**2)
        ldistind.extend(argwhere(temp<threshold)[:,1])
        ldistptr.append(len(ldistind))

        temp.sort()
        if(temp[0,n-s]!=0): lsigma.append(1/temp[0,n-s])
        else: lsigma.append(1)
            
            
    distptr=array(ldistptr)
    distind=array(ldistind)
    distv=array(ldist)

    dist=sparse.csr_matrix((distv,distind,distptr),shape=(n,n))

    sigma = sparse.dia_matrix((lsigma,0), shape=(n,n))
    
    print('1/6 done')
    return sigma, dist
```

```python
#get the Guassian kernel matrix

def getK(sigma, dist):
    K = -sigma.dot(dist.dot(sigma))
    K.data = exp(K.data)
      
    print('2/6 done')
    return K
```

```python
#Q is the matrix D^{-1}KD^{-1}

def getQ(K):
    D = sparse.dia_matrix((1/array(sqrt(sum(K,axis=1)))[:,0], 0), shape=(n,n))
    Q = D.dot(K.dot(D))
    
    print('3/6 done')
    return D, Q
```

```python
#get eigenvalues and eigenvectors of diffusion matrix

def get_eig(Q, D):
    [U, S, V] = svds(Q, dim)
    
    S = diag(flip(S))
    U = U[:,::-1]
    
    psi = D.dot(U)

    print('4/6 done')
    return S, psi
```

```python
#get diffusion map

def getY(S, psi):
    
    print('5/6 done')
    return psi.dot(S)
```

```python
#get labels

def getlabels(Y):
    
    kmeans = KMeans(clusters)
    kmeans = kmeans.fit(Y)
    labels = array(kmeans.predict(Y)).reshape(n,1)
    
    print('6/6 done')
    return labels
```

### 5. Calculate labels

```python
clusters = 10 #reduce image into how many colors
dim = 3       #embedding dimension
s = 7         #local scaling parameter

datav = imc.reshape(n, z)
```

```python
labels = diffuse(datav, clusters, dim)
```

### 6. Display result

```python
#calculate colors
colorrgb=[]
for i,j in product(range(clusters), range(z)):
    colorrgb.append(percentile( datav[argwhere(labels==i)[:,0],j],50 ))
    
    
#show image
imc=zeros((n,z))
for i in range(clusters): imc[argwhere(labels==i)]=colorrgb[3*i:3*i+3]
imc = imc.reshape(x,y,z)

plt.imshow(imc)
plt.show()
```
